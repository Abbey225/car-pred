```python
# transforming the data categorical variables using label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder(
)
```

```python
df.head(1)
```

```python
df.seller=le.fit_transform(df.seller)
df.fuel=le.fit_transform(df.fuel)
df.transmission=le.fit_transform(df.transmission)
df.owner=le.fit_transform(df.owner)
df.mileage=le.fit_transform(df.mileage)
df.max_power=le.fit_transform(df.max_power)
df.torque=le.fit_transform(df.torque)
```

```python

df.head(1)
```

```python
x=df.drop('selling_price',axis=1)
y=df.selling_price
```

```python
df.to_csv('carprice.csv')
```

```python

from sklearn.model_selection import train_test_split
```

```python
#spliting the dataset into train and test
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
et_model=ExtraTreesRegressor(n_estimators=750,random_state=100,min_samples_leaf=.0003)
```

```python
# training the model using extratrees regressor
et_model.fit(x_train,y_train)
```

```python
print(et_model.score(x_train,y_train))
print(et_model.score(x_test,y_test))
```

```python
from sklearn.metrics import r2_score
```

```python
#intitializing random forest regressor
rf_model=RandomForestRegressor(n_estimators=750)
```

```python
#fiting and training the model with random forest regressor
rf_model.fit(x_train,y_train)
```

```python
print(rf_model.score(x_train,y_train))
print(rf_model.score(x_test,y_test))
```
